```python
import pandas as pd
import numpy as np
First_Artist =pd.read_excel(r'D:\\TRANSFER EVERYONE\\LOP@\\praat music\\Conference paper\\AAOGE_JAB.xlsx',sheet_name='RAW')
First_Artist.head()


#PITCH VALUE COLUMN IS CONVERTED INTO AN ARRAY
Pitch_value_array=First_Artist.loc[:,'Pitch value']
PV_array=Pitch_value_array.values
print(PV_array)

PV_array=np.append(PV_array,[2000])

len(PV_array)





#code to generate the column frequency count
Seq_count=[]
count=1
for i in range(len(PV_array)-1):
    if (PV_array[i]==PV_array[i+1]):
        
        count=count+1
    else:
        print(count)
        Seq_count.insert(i,count)
        count=1


from itertools import repeat

Frequency_count=[]
for x in Seq_count:
    
    piv=int(x)
    for j in range(piv):
        Frequency_count.append(piv)
    

print(Frequency_count)
First_Artist["Frequency count"]=Frequency_count
First_Artist.head()







#code to generate the column sequence count

Sequence_count=[]
count=1
for i in range(len(PV_array)-1):
    if (PV_array[i]==PV_array[i+1]):
        Sequence_count.insert(i,count)
        
        count=count+1
    else:
        print(count)
        Sequence_count.insert(i,count)
        count=1

print(Sequence_count)
First_Artist["Sequence count"]=Sequence_count
First_Artist.head()




import os
from openpyxl import load_workbook

with pd.ExcelWriter('D:\\TRANSFER EVERYONE\\LOP@\\praat music\\Conference paper\\AAOGE_JAB.xlsx',engine='openpyxl',mode='a') as writer:
    First_Artist.to_excel(writer,sheet_name='FINAL_DATA',index=False)
```
